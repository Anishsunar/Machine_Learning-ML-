Dependent Events


*   Bag of Marbels(3 oranges and 2 yellow). Probablility of first removing a orange marbel?P(y)= P(y/o)= 2/4




P(o and y) = P(o)+P(y/o)-->Conditional Probablitty

P(A and B) = P(A) +P(B/A)


*   P(A and B) = P(B and A)

*   P(A)* P(B/A) = P(B)*P(A/B)


*   P(A/B) = (P(A)*P(B/A)) / P(B)



Variants:


*   Bernouli Naive Bayes: Whenever our features are in the form of 0 or 1.

*   Multinomial Naive Bayes: When input is in text.Dataset: Sentiment Analysis
*   Gaussian Naive Bayes: Continuous features(Age,Height,Weight,Overweight)



```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
X,y = load_iris(return_X_y = True, as_frame = True)
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
gnb = GaussianNB()
bnb = BernoulliNB()
mnb = MultinomialNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
bnb.fit(X_train,y_train)
```

```python
mnb.fit(X_train,y_train)
```

```python
y_pred_gnb = gnb.predict(X_test)
y_pred_bnb = bnb.predict(X_test)
y_pred_mnb = mnb.predict(X_test)
```
