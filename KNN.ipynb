```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import make_classification
X,y = make_classification(
    n_samples=1000, #1000 observation
    n_features = 3,# total independent features
    n_redundant= 1,
    n_classes = 2, # binary target /label
    random_state = 42

    )
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNeighborsClassifier()
```

```python
classifier = KNeighborsClassifier()
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
#GridSearchCV
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
```

```python
parameters = {
    'n_neighbors':list(range(30)),
    'p': [1,2],
    'weights': ['uniform','distance']

}
```

```python
clf = GridSearchCV(classifier,param_grid = parameters)
```

```python
# Splitting of train data to validation data
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
classifier = KNeighborsClassifier(n_neighbors = 25,p =1)
```

```python
classifier.fit(X_train, y_train)
```

```python
#Prediction
y_pred = classifier.predict(X_test)
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
# Import pickle for saving and loading the model
import pickle
# Save the trained model to a pickle file
with open('knn.pkl','wb') as file:
  pickle.dump(classifier,file)
print(f"Model saved to knn.pkl")
```

```python
# Load the model from the pickel file
with open('knn.pkl','rb') as file:
  loaded_model = pickle.load(file)
```

```python
# Test the loaded model with predictions
y_pred_loaded = loaded_model.predict(X_test)
print(y_pred_loaded)
```
